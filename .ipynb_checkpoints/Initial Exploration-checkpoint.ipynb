```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-darkgrid')
plt.rcParams['figure.figsize'] = [14, 6]
from scipy import stats
```

```python
# Read historic data, only working with QQQ for now
df = pd.read_csv("./historic_data.csv")
df["QQQ_perc_chng"] = df['QQQ_perc_chng'].str.replace('%', '').astype('float')
df.drop(["VTI_perc_chng", "VTI_price"], axis=1, inplace=True);
```

```python
print(df.head())
print(df.tail())
print(df.shape)
```

```python
# Descriptibve stats
print(df["QQQ_perc_chng"].describe())
```

```python
# Plot distribution
bins = 12
mean = df["QQQ_perc_chng"].describe()["mean"]
std = df["QQQ_perc_chng"].describe()["std"]
fig, ax = plt.subplots()
ax.hist(df["QQQ_perc_chng"], bins=bins, alpha=0.5, edgecolor="black")
#plt.xticks(rotation=30)
plt.title("QQQ Monthly return Distribution", loc='left', fontsize=14, fontweight=0, color='black')
plt.xlabel("Monthly % Return", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
ax.get_xaxis().set_visible(False)
# Draw mean, +- std. dev
plt.axvline(mean, color='blue', linestyle='dashed', linewidth=2, alpha=0.75)
plt.axvline(mean+std, color='blue', linestyle='dashed', linewidth=2, alpha=0.5)
plt.axvline(mean-std, color='blue', linestyle='dashed', linewidth=2, alpha=0.5)
plt.axvline(mean+2*std, color='blue', linestyle='dashed', linewidth=2, alpha=0.25)
plt.axvline(mean-2*std, color='blue', linestyle='dashed', linewidth=2, alpha=0.25)
plt.text(mean-0.5,-2,round(mean,2),rotation=0, fontsize=12)
plt.text(mean+std-0.75,-2,round(mean+std,2),rotation=0, fontsize=12)
plt.text(mean-std-0.75,-2,round(mean-std,2),rotation=0, fontsize=12)
plt.text(mean+2*std-0.75,-2,round(mean+2*std,2),rotation=0, fontsize=12)
plt.text(mean-2*std-0.75,-2,round(mean-2*std,2),rotation=0, fontsize=12)

plt.show()
```

```python
# Basic plotting
# Monthly returns of QQQ
fig, ax = plt.subplots()
ax.bar(df.index, df["QQQ_perc_chng"])
plt.xticks(rotation=30)
plt.title("QQQ Monthly return (2002-2020)", loc='left', fontsize=14, fontweight=0, color='black')
plt.xlabel("Month", fontsize=12)
plt.ylabel("QQQ % Change", fontsize=12)
plt.show()
```

```python
# Quick moving window analysis
# Returns a product of returns over a k month period

def window_prod(list, k):
    prod_arr = []
    for i in range(len(list)-k+1):
        prod = 1
        for j in range(k):
            #print("NEW RET: " + str(list[j]))
            prod = prod * (1 + (list[i+j]/float(100)))
            #print(prod)
        prod_arr.append(prod)
    return prod_arr
```

```python
# Quick tests @ 12, 18, 24, 30 month windows

k_list = [12, 18, 24, 30]
data_list= df["QQQ_perc_chng"].tolist()
for i in k_list:
    prod_arr = window_prod(data_list, i)
    print("Window Length: " + str(i) + " months")
    print(stats.describe(prod_arr))
    print("Median: " + str(np.median(prod_arr)) + "\n")
```
